## Activation Functions
This exercise is about the importance of selecting an appropriate activation function. Activation functions are crucial for introducing non-linearity into neural architectures. Each activation function has its unique characteristics and trade-offs and can significantly influence the model’s predictive capability, impacting its performance, convergence behavior, and the complexity of tasks it can effectively address.
This exercise explores activation functions for:
**Classification:**
- Sigmoid: Commonly used in binary classification tasks, the sigmoid function maps input values to a number between 0 and 1. However, it can suffer from so-called vanishing gradient problems, occurring in deeper networks.

- Hyperbolic Tangent (tanh): maps inputs to the range between -1 and 1 and addresses some of the shortcomings of the sigmoid function in terms of vanishing gradients.

- Softmax: is a generalization of the logistic function, commonly used for the output layer of multi-class classification networks, converting raw inputs into probabilities across multiple classes.


**Regression:**
- Rectified Linear Unit (ReLU): ReLU introduces non-linearity while maintaining computational efficiency. It helps mitigate vanishing gradient issues but may encounter "dead neurons" due to zero gradients for negative inputs.

- Leaky ReLU: addresses the "dying ReLU" problem by allowing a small, non-zero gradients for negative inputs, maintaining active neurons during training.

- Exponential Linear Unit (ELU): similar to ReLU for positive inputs, ELU applies an exponential function to negative values, ensuring smoother gradients and reducing the risk of “dead neurons” during trainin

```python
import torch

def linear(x):
    return x

def sigmoid(x):
    return 1 / (1 + torch.exp(-x))

def relu(x):
    return torch.maximum(x, torch.tensor(0.0, dtype=x.dtype, device=x.device))

def leaky_relu(x, alpha=0.01):
    return torch.where(x > 0, x, alpha * x)

def tanh(x):
    return torch.tanh(x)

def elu(x, alpha=1.0):
    return torch.where(x >= 0, x, alpha * (torch.exp(x) - 1))
```

**Classification**

  <div class="message-header">Loss functions</div>
  <div class="message-body">

  Let:
- $ y_i $ be the true value of the $i$-th sample in the training set, 
- $ N $ be the number of samples.

Define:
**Mean Squared Error (MSE)**

$$
\text{MSE}(y, \hat{y}) = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2
$$
where $ \hat{y}_i $ is the predicted value of the $i$-th sample in the training set.
**Mean Absolute Error**

$$
\text{MAE}(y, \hat{y}) = \frac{1}{N} \sum_{i=1}^{N} |y_i - \hat{y}_i|
$$
where $ \hat{y}_i $ is the predicted value of the $i$-th sample in the training set.. 
**Binary Cross-Entropy Loss (BCE)**

$$
\text{BCE}(y, \hat{y}) = - \frac{1}{N} \sum_{i=1}^{N} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
$$
where $ \hat{y}_i $ is the predicted probability of the $i$-th sample in the training set.


  </div>
</article>


</div></article>



```python
# Custom Binary Cross-Entropy Loss Function
class BCE_Loss(nn.Module):
    def __init__(self):
        super(BCE_Loss, self).__init__()

    def forward(self, outputs, targets):
        epsilon = 1e-12
        outputs = torch.clamp(outputs, min=epsilon, max=1-epsilon)
        bce = -(targets*torch.log(outputs) + (1-targets)*torch.log(1-outputs))
        return torch.mean(bce)

            

# Custom Mean Squared Error Loss Function
class MSE_Loss(nn.Module):
    def __init__(self):
        super(MSE_Loss, self).__init__()

    def forward(self, outputs, targets):
        mse = (outputs - targets)**2
        return torch.mean(mse)


# Custom Mean Absolute Error Loss Function
class MAE_Loss(nn.Module):
    def __init__(self):
        super(MAE_Loss, self).__init__()

    def forward(self, outputs, targets):
        mae = torch.abs(outputs - targets)
        return torch.mean(mae)
```

```python
# Task 5.2: Explanation of why BCE is preferred for classification problems
# 1. BCE (Binary Cross-Entropy) is specifically designed for binary classification problems. It treats each prediction as
#    the probability of the positive class (y=1) and penalizes predictions that are far from the true labels. By minimizing
#    BCE, the model is encouraged to produce outputs close to 1 for the positive class and 0 for the negative class.
#
# 2. Unlike MSE or MAE, BCE accounts for the probabilistic nature of classification. Predictions closer to 0.5 (uncertain)
#    incur higher penalties in BCE, whereas MSE and MAE treat 0.5 as an "acceptable" mid-point prediction.
#
# 3. BCE implicitly enforces large gradients for poorly calibrated predictions (e.g., predicting 0.1 for y=1), enabling
#    faster updates to model parameters. This is particularly beneficial during training.
#
# 4. MSE and MAE, while computationally simpler, are not optimal for classification:
#    - MSE: Penalizes large errors quadratically, which can make training unstable. Also, MSE is symmetric, failing to
#           reflect the asymmetry in misclassifying positive and negative classes.
#    - MAE: Penalizes errors linearly but is less sensitive to small deviations, which can slow training convergence.
#
# In conclusion, BCE's ability to handle probabilistic outputs, its asymmetry, and its sensitivity to uncertainty make it
# the preferred loss function for binary classification tasks, as evidenced in the visualization where it reacts sharply
# to errors around the critical threshold of 0.5.
```

Consider the case of iris codes derived from an individual’s iris pattern. The codes are represented as binary vectors which is matched against a database of authorized codes to determine access. The Hamming Distance (HD) is commonly used as a similarity measure.

$$
HD(\mathbf{x}, \mathbf{y}) = \sum_{i=1}^{n}

1 & \text{if } x_i \neq y_i \\

It calculates the number of positions with different values.

However, this approach is not valid for a loss function in the context of neural network training.

```python
# Task 8: Why the Hamming Distance (HD) is not a valid choice as a loss function

# The Hamming Distance (HD) calculates the number of differing positions between two binary vectors, defined as:
# HD(x, y) = Σ_{i=1}^{n} 1 if x_i ≠ y_i, else 0
#
# While HD is effective as a similarity measure, it is not suitable as a loss function in neural network training for the following reasons:

# 1. **Non-Differentiable Nature**:
#    - HD uses a binary condition (1 if x_i ≠ y_i, else 0), which results in discrete, step-like changes.
#    - Gradient Descent, the optimization algorithm used to train neural networks, relies on the computation of
#      gradients (partial derivatives) of the loss function. Since HD is not continuous, it does not have well-defined
#      derivatives, making it incompatible with backpropagation.

# 2. **Lack of Gradient Information**:
#    - Even if we attempt to approximate gradients, the binary output of HD provides no useful gradient information
#      about how to adjust the weights to minimize the loss.
#    - For example, a small change in the model's parameters may have no effect on HD until a bit flips, resulting
#      in a sudden, non-smooth loss change.

# 3. **Optimization Challenges**:
#    - Neural network training aims to find a smooth loss landscape where gradients guide parameter updates efficiently.
#      HD's step-like behavior creates a flat loss landscape in most regions, making optimization inefficient or
#      impossible.

# 4. **Misalignment with Neural Network Outputs**:
#    - Neural network outputs are typically real-valued (continuous) during training, requiring loss functions that
#      can handle such outputs (e.g., BCE, MSE).
#    - HD, being defined only for binary vectors, does not align with this requirement and would require additional
#      transformations (e.g., thresholding), further complicating optimization.

# 5. **Gradient Descent and Continuous Functions**:
#    - Gradient Descent relies on continuous and smooth functions like BCE or MSE to iteratively reduce loss.
#    - These functions penalize errors proportionally and provide meaningful gradients even for small deviations,
#      ensuring efficient weight updates.

# In conclusion:
# The Hamming Distance is not valid as a loss function for neural network training due to its non-differentiable nature,
# lack of gradient information, and misalignment with continuous neural network outputs. A differentiable alternative
# such as Binary Cross-Entropy (BCE) or Mean Squared Error (MSE) should be used to optimize neural network models
# for tasks involving binary vector similarity.
```
