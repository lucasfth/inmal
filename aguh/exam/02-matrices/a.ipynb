## 1. Matrices

A matrix is a rectangular of size $m \times n$ for $m$ rows and $n$ columns. In machine learning, we can combine multiple vectors representing data points to create a model.

## 2. Matrix operations

### Addition

For two matrices of equal size, we can add or subtract each component such that we get a resulting matrix of same size.

### Scalar Multiplication

We can multiply a matrix with a single scalar such that we multiply each component of the matrix with the scalar value, essentially scaling the matrix.

### Matrix Multiplication

We can also multiply two matrices with each-other by the following formula:

$c_{ij} = \sum_{k=1}^n a_{ik} * b_{kj}$

for matrices A and B.

Given two matrices $n \times m$ and $m \times k$, such that the first matrix' number of columns equals the second matrix' number of rows, we can calculate each cell $c_{ij}$ in the resulting $n \times k$ matrix as:  
Start on the left-most cell of the $i$th row in $A$ and the top-most cell of the $j$th column in $B$ and multiply these values. Then multiply these two values and go one to the right and one down in respectively $A$ then $B$ and multiply these values. Repeat for a total of $m$ repetitions and sum all these multiplication results together to get the value for $c_{ij}$.

Matrix multiplication can be used to solve sets of linear equations.

Note that matrix multiplication is non-commutative meaning $AB \neq BA$.

### Transpose

We write $A^T$ to denote a transposed matrix, meaning it is flipped over its diagonal axis (axis from top left to bottom right). This is used to calculate covariance matrices.

$\begin{matrix}
    a & b \\
    c & d \\
\end{matrix}$

turns to

$\begin{matrix}
    a & c \\
    b & d \\
\end{matrix}$

### Determinant

The scalar value for the determinant of a square matrix is denoted $det(A)$ and is used to figure out if a matrix can be inverted (see following section). If the determinant is 0, no solution exists and the matrix is called singular.

### Inverse

The inverse of a matrix is written $A^{-1}$ and is defined as a matrix that makes $AA^{-1}=I$ true for the identity matrix $I$. The identity matrix is the square matrix (size $n \times n$) such that for component $c_{ij}$ where $i=j$, the component is $1$ and the remaining are $0$. There only exists an inverse matrix for square matrices with a non-zero determinant. The inverse is used to solving systems of linear equations.

For a $2 \times 2$ matrix
$A = [\begin{matrix}
    a & b \\
    c & d \\
\end{matrix}]$

the inverse matrix can be found as follows

$
A^{-1}=\frac{1}{ad-bc}
[\begin{matrix}
    d & -b \\
    -c & a \\
\end{matrix}]$

Two rules for inverses:

$(ABC)^{-1} = C^{-1}B^{-1}A^{-1}$  
$(A^T)^{-1} = (A^{-1})^T$

### Orthogonality

Two matrices are orthogonal if its transpose equals its inverse: $A^T = A^{-1}$. If it is multiplied by its tranposed self it will equal the identity matrix. It is used to rotate matrices.
