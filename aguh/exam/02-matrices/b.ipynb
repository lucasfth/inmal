Focus on how linear equations and their solutions are related to matrices.

## Linear equations

A set of $n$ linear equations with $n$ unknowns can be solved by combining matrix multiplication, matrix determinants and inverse matrices.

We can write a set of linear equations as $Ax = b$ where $A$ is the coefficient matrix, $x = n \times 1$ vector of variables and $b = n \times 1$ vector of constants for a matrix $A$ of size $n \times n$ (square).

In other words, for each equation create a row in $A$ such that $a_1x_1 + ... + a_nx_n + b$ for variables $x_1, ..., x_n$ and constant $b$ is rewritten to the row in $A$ as $[a_1, ..., a_n]$. Each column should of course be aligned so it represents the same variable $x_i$. The 1-column vector $b$ then consists of the constants from each linear equation in the same row as the coefficients are in $A$.

We can then rewrite $Ax=b$ to $AA^{-1}x=bA^{-1} => x=bA^{-1}$ for invertible matrices $A$.

This means we find the inverse matrix $A^{-1}$ and use matrix multiplication with the constants $b$ we get the values in a 1-column vector for $x$ such that the first row in $x$ corresponds to $x_1$ and the last row corresponds to $x_n$.

Example:

$3x_1 + 4x_2 = 35$  
$4x_1 + 2x_2 = 20$  

Note that for linear equations we can rewrite them to the form $a_1x_1 + ... + a_n+x_n = b$ - if not, the equation is not linear (e.g. it contains $x_i * x_j$ for two input variables $x_i,x_j \in \{x_1 ... x_n\}$). Note that the above linear equations might as well have used $x$ and $y$ as input variables than $x_1$ and $x_2$, but for simplicity we do the latter.

We can then rewrite them to the form $Ax = b$ as follows:

$A = [
\begin{matrix}
    3 & 4 \\
    4 & 2 \\
\end{matrix}]$  

$x =
[\begin{matrix}
    x_1 \\
    x_2 \\
\end{matrix}]$  

$b = 
[\begin{matrix}
    35 \\
    20 \\
\end{matrix}]$  

The inverse of A can be found as follows (since it is a 2x2 matrix):

$A^{-1}=\frac{1}{ad-bc}
[\begin{matrix}
    d & -b \\
    -c & a \\
\end{matrix}]$  

where $det(A) = ad-bc$ and if this value is 0 (we cannot divide by 0) there does not exist a solution.

$A^{-1}=\frac{1}{2*3-16}
[\begin{matrix}
    2 & -4 \\
    -4 & 3 \\
\end{matrix}] = 
[\begin{matrix}
    -0.2 & 0.4 \\
    0.4 & -0.3 \\
\end{matrix}]$  

We can now do matrix multiplication for $A^{-1} \cdot b$ to calculate $x$:  

$x =
[\begin{matrix}
    -0.2 & 0.4 \\
    0.4 & -0.3 \\
\end{matrix}]
\cdot
[\begin{matrix}
    35 \\
    20 \\
\end{matrix}] =
[\begin{matrix}
    -0.2 * 35 + 0.4 * 20 \\
    0.4 * 35 - 0.3 * 20 \\
\end{matrix}] =
[\begin{matrix}
    1 \\
    8 \\
\end{matrix}]$  

Since $x = [\frac{x_1}{x_2}]$ we have that $x_1 = 1$ and $x_2 = 8$.
